Sampel 1000 data produk<br>
<ol>
    <li>Kategori Produk: Elektronik, Pakaian, Makanan, Minuman</li>
    <li>Brand: BrandA, BrandB, BrandC, BrandD</li>
    <li>Harga: 5000 - 2.000.000</li>
    <li>Stok: 0 - 100</li>
    <li>Diskon: 0%, 10%, 20%, 30%</li>
    <li>Status: Tersedia, Habis</li>
    <li>Lokasi Gudang: Jakarta, Bandung, Surabaya</li>
    <li>Rating: 1 - 5</li>
    <li>Penjualan Bulanan: 0 - 1000</li>
    <li>Ketersediaan Online: Ya, Tidak</li>
</ol>

```python
import numpy as np
import pandas as pd

n_data = 1000

kategori = np.random.choice(['Elektronik', 'Pakaian', 'Makanan', 'Minuman'], size=n_data)
brand = np.random.choice(['BrandA', 'BrandB', 'BrandC', 'BrandD'], size=n_data)
harga = np.random.randint(5000, 2000001, size=n_data)
stok = np.random.randint(0, 101, size=n_data)
diskon = np.random.choice([0, 10, 20, 30], size=n_data)
status = np.where(stok > 0, 'Tersedia', 'Habis')
lokasi = np.random.choice(['Jakarta', 'Bandung', 'Surabaya'], size=n_data)
rating = np.random.randint(1, 6, size=n_data)
penjualan = np.random.randint(0, 1001, size=n_data)
online = np.random.choice(['Ya', 'Tidak'], size=n_data)

df = pd.DataFrame({
    'Kategori': kategori,
    'Brand': brand,
    'Harga': harga,
    'Stok': stok,
    'Diskon (%)': diskon,
    'Status': status,
    'Lokasi Gudang': lokasi,
    'Rating': rating,
    'Penjualan Bulanan': penjualan,
    'Online': online
})


print(df.info())
print(f"Harga termurah: {df.Harga.min()}")
print(f"Harga termahal: {df.Harga.max()}")
print(f"Rata-rata harga: {df.Harga.mean():,.2f}")

print(f"Jumlah produk Elektronik: {df[df['Kategori']=='Elektronik'].shape[0]}")
print(f"Jumlah produk Pakaian   : {df[df['Kategori']=='Pakaian'].shape[0]}")
print(f"Jumlah produk Makanan   : {df[df['Kategori']=='Makanan'].shape[0]}")
print(f"Jumlah produk Minuman   : {df[df['Kategori']=='Minuman'].shape[0]}")

print(f"Jumlah produk habis: {df[df['Status']=='Habis'].shape[0]}")
print(f"Jumlah produk tersedia: {df[df['Status']=='Tersedia'].shape[0]}")

print(f"Jumlah produk online: {df[df['Online']=='Ya'].shape[0]}")
print(f"Jumlah produk offline: {df[df['Online']=='Tidak'].shape[0]}")

print(f"Rata-rata rating: {df.Rating.mean():.2f}")
print(f"Total penjualan bulanan: {df['Penjualan Bulanan'].sum()}")


df.to_csv('dataProduk.csv', index=False)
```
